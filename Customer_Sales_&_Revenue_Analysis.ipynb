```python
import pandas as pd
```

```python

data = {
    "Customer_ID": [101, 102, 103, 101, 104, 102, 105],
    "Customer_Name": ["Amit", "Ravi", "Neha", "Amit", "Sonal", "Ravi", "Pooja"],
    "Product": ["Door Handle", "Lock", "Kitchen Trolley", "Lock", "Door Handle", "Locker", "Bathroom Accessory"],
    "Month": ["Jan", "Jan", "Feb", "Mar", "Mar", "Apr", "Apr"],
    "Sales_Amount": [5000, 7000, 8000, 6000, 9000, 12000, 4000]
}

df = pd.DataFrame(data)
df
```

```python
total_sales = df["Sales_Amount"].sum()
avg_sales = df["Sales_Amount"].mean()

print(f"Total Sales: {total_sales}")  # Total revenue generated
print(f"Average Sales: {avg_sales}")  # Average transaction value
```

```python
monthly_sales = df.groupby("Month")["Sales_Amount"].sum()
monthly_sales
```

```python
top_customers = df.groupby("Customer_Name")["Sales_Amount"].sum().sort_values(ascending=False)
print(top_customers)
```

```python
#  VisVisualization
import matplotlib.pyplot as plt
import seaborn  as sns
```

```python
fig, ax = plt.subplots(1, 2, figsize=(7,5))

sns.barplot(data=monthly_sales, ax=ax[0])
ax[0].set_title("Monthly Sales Revenue")
ax[0].set_xlabel("Month")
ax[0].set_ylabel("Sales Amount")

sns.barplot(data=top_customers,  ax=ax[1])
ax[1].set_title("Top Customers by Revenue")
ax[1].set_xlabel("Customer Name")
ax[1].set_ylabel("Sales Amount")

fig.tight_layout()
```
